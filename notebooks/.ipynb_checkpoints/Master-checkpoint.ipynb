```python
# import packages for alpaca, panadas and OS/dotenv to read .env file with api keys
import os
import pandas as pd
import numpy as np
import alpaca_trade_api as tradeapi
# Load .env enviroment variables
from dotenv import load_dotenv
load_dotenv()
import pathlib as Path
```

```python
#helpful maybe?
from datetime import datetime, timedelta

days = 90
startdate = datetime.now() - timedelta(days=days)

# for fixed dates
start = pd.to_datetime('2002-01-01')
end = pd.to_datetime('2020-01-01')
```

```python
# storing alpaca key and secret key in variables 
alpaca_api_key = os.getenv('ALPACA_API_KEY')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')

api = tradeapi.REST(alpaca_api_key,alpaca_secret_key,api_version='V2')
```

```python
# set time frame
timeframe = '1D'

# start and 'end' date for covid stock market 


start_date_covid = pd.Timestamp('2019-12-01',tz='America/New_York').isoformat()
end_date_covid = pd.Timestamp('2020-11-20',tz='America/New_York').isoformat()

# start and end date for 2008 recession 
start_date_2008 = pd.Timestamp('2007-07-01',tz='America/New_York').isoformat()
end_date_2008 = pd.Timestamp('2009-07-01',tz='America/New_York').isoformat()

# set entertainment tickers (per industry)
ent_tickers = ['IMAX','DIS','NFLX','CNK','SNE']
fast_food_tickers = ['MCD','WEN','YUM','DPZ','CMG']
# financeind_tickers
# techind_tickers etc

# get stock data from covid
entertainment_tickers_covid  = api.get_barset(ent_tickers,timeframe,start=start_date_covid,end = end_date_covid).df
fastfood_tickers_covid = api.get_barset(fast_food_tickers,timeframe,start=start_date_covid,end = end_date_covid).df


# get stock data from 2008 recession
entertainment_tickers_2008 = api.get_barset(ent_tickers,timeframe,start=start_date_2008,end = end_date_2008).df
fastfood_tickers_2008 = api.get_barset(fast_food_tickers,timeframe,start=start_date_2008,end = end_date_2008).df
```
